```python
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
print(sns.get_dataset_names()) ## list of all datasets that are in-built in Seaborn
```

```python
df = sns.load_dataset("car_crashes") 
df.head(5)
```

```python
plt.scatter(df.no_previous,df.alcohol)
#style the plot 
sns.set_style("white")
sns.despine()
plt.show()
```

```python
param=sns.axes_style()
param
```

```python
plt.scatter(df.no_previous,df.alcohol)
#style the plot 
sns.despine()
sns.set_context("notebook")
plt.show()
sns.set_style("darkgrid", {'grid.color': '.5'})
```

```python
sns.displot(df['not_distracted'], kde=False, bins=25)
```

```python
sns.kdeplot(df['alcohol'])
```

```python

sns.relplot(x='alcohol', y='total', hue=df.alcohol.tolist(), data=df);
```

```python
sns.barplot(x='total', y='alcohol', data=df);
```
